# Goal: Techniques to find the Correlation between different input features:

- Input: 'InputTableFeatures.csv'  # created file
- Output: Correlation matrix 

### Techniques used to find correlations:
1. Pandas corr() : using pearson, kendall and spearman methods
    - Using Matplotlib to add colour cells


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pdData = pd.read_csv('InputTableFeatures.csv')
```

```python
pdData
```

```python
pdData.columns
```

```python
InputFeatures = ['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5',
       'Feature6', 'Feature7', 'Feature8', 'Feature9',
       'Feature10', 'Feature11']
```

```python
pdData[InputFeatures].describe()
```

# Pandas corr()

```python
pdData[InputFeatures].corr(method ='pearson') #pearson : standard correlation coefficient
```

```python
pdData[InputFeatures].corr(method ='kendall') #Kendall Tau correlation coefficient
```

```python
pdData[InputFeatures].corr(method ='spearman') #Spearman rank correlation
```

# Using Matplotlib to add colour cells

```python
pdData[InputFeatures].corr(method='pearson').\
            style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
pdData[InputFeatures].corr(method='kendall').\
            style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
pdData[InputFeatures].corr(method='spearman').\
            style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```
